```python
# Import the libraries to work on
import pandas as pd
import dateutil
import math

# Read the dataset to work on

datasets = pd.read_csv('/home/uk/Documents/Assignments and projects/Beta calculation/index.csv', error_bad_lines=False)
print ('No of observations in the dataset: ' + str(len(datasets)))

# Find the data type of the variable datasets
print('Type of datasets: ' + str(type(datasets)))

# Now perform the cleaning of dataset
datasets.head()

# Remove the empty fields in the different columns
datasets['duration (seconds)'].fillna(0) # Filling out the empty fields with 0

# Check if above step is done perfectly

# Because this column is in str we need to convert it into int
duration_in_int = pd.to_numeric(datasets['duration (seconds)'], errors='coerce')



print('MEAN of Duration in (seconds): ' + str(duration_in_int.sum() / len(duration_in_int)))
print('MEAN of Duration in (minutes): ' + str((duration_in_int.sum() / len(duration_in_int))/60))
print('MEAN of Duration in (hours): ' + str(((duration_in_int.sum() / len(duration_in_int)/60)/60))) 


# Find the Distinct locations where these sightings have been made

# where there are empty fields, we replace it with unknown locations
datasets.city.fillna('Unknown Location') 
datasets.city.str.upper()
datasets.city.str.strip()

keys = datasets.groupby(['city']).groups.keys()

print('Total cities where there have been sightings: ' + str(len(keys)))
print('Average sightings per city: ' + str(math.ceil(len(datasets)/len(keys))))

print('\n\n')

print(str(type(keys)))

list(keys)
```
